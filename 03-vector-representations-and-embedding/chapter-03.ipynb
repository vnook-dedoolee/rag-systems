# Глава 3. Векторные представления и эмбеддинги

## 3.1. Принципы работы с векторными представлениями

Раньше информацию сортировали по алфавиту или темам. Векторный подход работает иначе: он размещает слова и тексты в пространстве, опираясь только на их смысл.

### Что такое векторные представления

Эмбеддинг (векторное представление) превращает текст в числа. Но это не просто код букв, а перевод смысла на язык математики. Каждое слово становится точкой в пространстве со множеством измерений. Положение точки отражает значение слова.

Главное свойство: близкие по смыслу слова находятся рядом. «Кот» и «кошка» — соседи, «радость» и «счастье» — тоже. А «автомобиль» и «яблоко» — далеко друг от друга.

Размерность — это количество координат у каждого слова. Современные модели используют от 128 до 4096 измерений. Человек такое пространство представить не может, но математика работает с ним легко.

### Дистрибутивная гипотеза: основа всего

В основе векторных представлений лежит простая, но мощная идея: «скажи мне, с кем ты водишься, и я скажу, кто ты». В лингвистике это называется дистрибутивной гипотезой — слова, встречающиеся в схожих контекстах, имеют схожие значения.

Языковая модель анализирует миллиарды предложений и замечает закономерности: после слова «красивый» часто идут «дом», «цветок», «закат», а после «быстрый» — «автомобиль», «бег», «интернет». Постепенно формируется понимание того, какие слова «дружат» друг с другом.

Контекст становится учителем — модель не знает, что означает слово «коса», но видит, что иногда рядом с ним появляются «волосы», «причёска», «парикмахер», а иногда — «сено», «покос», «косарь» (а есть ещё и третий контекст). Так одно слово получает несколько векторных представлений в зависимости от контекста.

### Семантическое пространство: география смыслов

Векторное пространство — это как карта понятий. Животные группируются в одной области, транспорт — в другой, эмоции — в третьей. Расстояние между точками показывает, насколько смыслы близки.

Главная мера близости — косинусное сходство. Оно оценивает, насколько векторы смотрят в одну сторону, и не учитывает их длину. Если показатель близок к 1 — слова почти синонимы. Если к 0 — связи между ними нет. Так удобно, потому что косинус равен 1 при угле 0 градусов и 0 при прямом угле в 90 градусов.

### Косинусное сходство векторов‑эмбеддингов



На графиках видны примеры:
- «кот» и «кошка»: угол 7,9°, сходство почти 1 (0,991) — слова близки.
- «кот» и «математика»: угол 72,7°, сходство 0,298 — слова далеки друг от друга.
Чем меньше угол, тем ближе смысл.

Удивительно, что со смыслами можно работать как с числами. Знаменитый пример:
вектор «король» – вектор «мужчина» + вектор «женщина» ≈ вектор «королева».
Математика оперирует понятиями напрямую.

### Фундаментальные принципы работы с векторами


1. **Единый размер.** Слово, предложение или целый документ — всё превращается в вектор одной длины. «Да» и длинный абзац получат векторы одинакового размера. Поэтому можно сравнивать любые куски текста.

2. **Плотность.** В векторных представлениях каждая координата что-то значит и может быть любым числом. В отличие от разреженных векторов, где почти везде нули, здесь каждое измерение хранит часть смысла.

3. **Контекстуальность.** Современные модели учитывают окружение слова. «Ключ» от замка и «ключ» как родник получат разные векторы, потому что находятся в разном контексте.

4. **Сложение смыслов.** Векторы можно объединять. Вектор целого предложения получают через усреднение векторов слов или более сложные способы с учётом грамматики и смысла.

5. **Устойчивость.** Небольшие изменения текста не должны сильно менять вектор. Опечатки, синонимы или пересказ другими словами должны давать похожие векторы, чтобы сохранялась смысловая целостность.

### Практические следствия принципов

**Поиск по смыслу.** Тексты, близкие по смыслу, оказываются рядом в векторном пространстве. Запрос «как починить автомобиль» найдёт документы про «ремонт машины», даже если слова в них не совпадают.

**Группировка и рубрикация.** Тексты одной темы сами собираются в одной области. Алгоритмы могут использовать это для автоматической сортировки документов.

**Рекомендации.** Векторные представления помогают находить похожие товары, фильмы или статьи. Если пользователю что-то понравилось, система ищет векторы с близкими координатами.

**Итог.** Векторные представления изменили работу с текстом: вместо возни с символами мы работаем с геометрией смыслов. Благодаря этому появились современные поисковики, переводчики, чат-боты и системы, которые по-настоящему «понимают» язык через математику многомерных пространств.

## 3.2. Векторные базы данных и их применение

Векторные представления — это язык смысла для машин, а векторные базы данных — библиотеки, где эти смыслы хранятся. В отличие от обычных баз, которые ищут точные совпадения, векторные находят похожее и близкое — именно это нужно для интеллектуального поиска.

### Принципы устройства векторного хранилища

Векторная база данных — это хранилище, оптимизированное под многомерные числовые массивы. Вектор — точка в пространстве из сотен измерений, а задача системы — быстро находить ближайших соседей.

Индексирование здесь работает иначе, чем в обычных базах. Вместо B-деревьев и хеш-таблиц применяют алгоритмы приближённого поиска: HNSW (Hierarchical Navigable Small World), IVF (Inverted File), LSH (Locality Sensitive Hashing). Они жертвуют точностью ради скорости — для семантического поиска это оправдано.

Ключевую роль играют метаданные. Каждая запись, помимо вектора, содержит исходный текст, категорию, дату или права доступа. Это позволяет сочетать семантический поиск с обычной фильтрацией.

### Современные решения: ландшафт векторных баз данных

**Chroma** — решение с открытым кодом для стартапов. Философия — «запустить за 5 минут», без сложного администрирования. Хороша для аудио и мультимодального контента, легко интегрируется с ML-фреймворками. Минус — ограниченная масштабируемость.

**Pinecone** — закрытое облачное решение, лидер среди управляемых сервисов. Сложность инфраструктуры скрыта за простым API. Поддерживает real-time обновления, фильтрацию метаданных, интеграцию с OpenAI, AWS, Google Cloud. Минусы — высокая цена и зависимость от вендора.

**Qdrant** — высокопроизводительная система с открытым кодом на Rust. Быстрый поиск, низкое потребление памяти, поддержка сложных фильтров и гибридного поиска. Доступна локально и в облаке. Минус — документация и экосистема слабее, чем у конкурентов.

**Weaviate** — объединяет векторный поиск с графовой базой данных. Объекты связаны по смыслу и структуре. Встроенные модули извлекают векторы из текста, изображений и аудио, GraphQL API даёт гибкость запросов. Минус — сложность для простых проектов.

**Milvus** — для петабайтов векторных данных. Распределённая архитектура с горизонтальным масштабированием, богатый выбор алгоритмов индексирования, интеграция с Kubernetes и облаками. Минус — требует высокой экспертизы.

**pgvector** — расширение, превращающее PostgreSQL в векторную базу. Для команд, уже сидящих на Postgres, — естественный путь без изучения новых технологий. Сохраняются ACID, бэкапы, мониторинг. Минус — уступает специализированным системам по производительности.

### Критерии выбора: навигация в многообразии

Выбор векторной базы зависит от масштаба проекта, экспертизы команды, требований к производительности и бюджета. Универсального решения нет — каждая система оптимизирована под свои сценарии.

## 3.3. Сегментирование текстов как основа RAG‑сервисов

Векторные представления — это язык смыслов. Сегментирование текста (чанкинг) — это членораздельное произношение на этом языке. От разбивки документа зависит, найдёт ли RAG-система информацию и даст ли осмысленный ответ.

### Фундаментальная проблема сегментирования

Сегментирование — это баланс между детализацией и контекстом. Слишком мелкие фрагменты теряют смысл, слишком крупные — не влезают в окно внимания модели и снижают точность. По данным исследований, качество чанкинга меняет производительность RAG-системы на 10–15%.

Основная дилемма: оптимальный размер чанка зависит от типа документа, предметной области, характера запросов и модели эмбеддингов. Финансовые отчёты требуют иного подхода, чем художественная литература, техническая документация — иного, чем новостные статьи.

### Метод фиксированного размера: простота как преимущество

Самый распространённый подход — фиксированный размер фрагментов. Документ нарезается на куски по 256, 512 или 1024 токена без учёта содержания. Метод привлекает простотой и предсказуемостью.

Главные плюсы — скорость и единообразие. Одинаковая длина фрагментов упрощает индексирование и поиск. Для больших корпусов однородных документов это оптимальный вариант.

Критическая проблема — разрыв смысловых границ. Предложение режется пополам, абзац разрывается, таблица теряет структуру. Система ищет ответ во фрагменте, где есть лишь половина информации.

### Скользящее окно: сохранение контекста через перекрытие

Метод скользящего окна решает проблему потери контекста через перекрытие фрагментов. Каждый новый чанк начинается с отступом назад, создавая зону перекрытия.

Типичная схема: размер чанка 500 токенов, перекрытие — 100. Первый фрагмент — токены 1–500, второй — 400–900, третий — 800–1300. Это гарантирует, что информация не разорвётся между границами.

Главный плюс — контекстуальная связность. Ключевая информация на границе фрагментов обязательно попадёт целиком в один из перекрывающихся чанков. Это важно для сложных рассуждений и многоэтапных объяснений.

Недостаток — рост объёма данных на 20–50% в зависимости от перекрытия. Избыточность замедляет индексирование и увеличивает векторную базу. Одна и та же информация может появляться в выдаче несколько раз.

### Семантическое сегментирование: следование логике текста

Семантический чанкинг делит текст по смысловым границам — конец темы, смена аспекта, новый аргумент, — а не по количеству символов.

Современные подходы используют языковые модели: каждое предложение анализируется в контексте соседних, чтобы найти границу логического блока. Применяются отслеживание внимания модели и метрики когерентности текста.

---
Когерентность текста — это смысловая связность, когда каждое следующее предложение логично вытекает из предыдущего, не противоречит ему и работает на общую тему, как звенья одной цепи.

---

Перплексия служит индикатором смысловых разрывов. Если модель с трудом предсказывает следующее предложение, это сигнал к смене темы и естественная точка разделения.

---
Перплексия (англ. Perplexity, PPL) — метрика для оценки качества языковых моделей в области обработки естественного языка. Она измеряет, насколько хорошо модель предсказывает последовательность слов или токенов в тестовом наборе данных. Интуитивно перплексию можно понимать как «эффективное» количество вариантов, из которых модель выбирает на каждом шаге. Если перплексия равна 100, это означает, что неопределённость модели эквивалентна выбору одного из 100 равновероятных исходов.

---

Результат — фрагменты переменной длины с высокой внутренней связностью. Каждый чанк содержит завершённую мысль, что улучшает качество поиска. Исследования показывают рост точности на 5–12% по сравнению с фиксированным делением.

Главное ограничение — вычислительная сложность. Анализ границ требует обращения к языковой модели и замедляет обработку в 5–10 раз.

### Структурное разделение: использование архитектуры документа

Структурный чанкинг опирается на естественную организацию документа: главы, разделы, абзацы, списки, таблицы становятся единицами сегментирования с сохранением иерархии.

Для HTML и Markdown структурное деление тривиально — заголовки, блоки кода и таблицы уже выделены разметкой. PDF требуют более сложного анализа макета.

Финансовые отчёты выигрывают от структурного подхода особенно. Таблицы с показателями, списки рисков, прогнозные разделы работают лучше как отдельные чанки, сохраняя внутреннюю логику и форматирование.

Иерархия позволяет создавать чанки разной детализации: заголовок раздела — отдельный фрагмент, подразделы — дочерние, связанные с родителем через метаданные.

### Адаптивное сегментирование: искусственный интеллект в помощь

Адаптивные методы обучаются оптимизировать разделение под конкретную задачу на примерах качественных чанков.

Поздний чанкинг обрабатывает документ целиком для получения контекстуальных эмбеддингов, а затем делит с учётом выявленных взаимосвязей.

Агентный чанкинг распределяет анализ между специализированными агентами: один отвечает за синтаксис, другой — за семантическую связность, третий — за соответствие предметной области.

Метачанкинг — наиболее продвинутый подход: создаётся иерархия фрагментов от предложений до абзацев. Динамическое объединение позволяет выбирать гранулярность под конкретный запрос.

### Критерии выбора метода сегментирования

**Факторы выбора:**

- **Тип контента.** Структурированные документы — структурный подход, художественные тексты — семантический, новостные сводки — фиксированное деление.
- **Вычислительные ресурсы.** Семантическое сегментирование затратно, фиксированное — почти мгновенно. Для больших корпусов это решающий фактор.
- **Характер запросов.** Фактологические вопросы — мелкие чанки, аналитические — крупные фрагменты с богатым контекстом.
- **Предметная область.** Юридические документы требуют сохранения структуры статей, научные — целостности методологии, технические — связности пошаговых инструкций.

**Лучший результат** часто дают гибридные подходы: структурное деление на верхнем уровне плюс семантическое внутри разделов, либо фиксированный размер с семантической корректировкой границ.

**Итог:** качество сегментирования определяет успех RAG-системы на 40–60%. Продуманная стратегия чанкинга окупается точностью поиска и релевантностью ответов. Это стратегический архитектурный выбор, а не техническая деталь.
